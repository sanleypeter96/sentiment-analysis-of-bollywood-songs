# Importing packages

```python
import numpy as np
import os
import io
import pandas as pd
import re
from os import listdir
from os.path import isfile, join
```

# Reading the first dataset

```python
os.getcwd()
data_folder = os.path.join(os.getcwd(),'../Dataset/bolly_song_lyrics/scraped_lyrics_cleaned/')  # path to raw song files
```

```python
file_temp = [os.path.join(data_folder,i) for i in sorted(os.listdir(data_folder)) if os.path.isfile(os.path.join(data_folder,i))]

df = pd.DataFrame()

#for every file in folder, read it and append to a empty dataframe with column filename as 'Title'
for file in file_temp:
    print(file)
    _df = pd.read_csv(file, names =['lyrics'], header=None, index_col=False, sep='\\n', encoding='utf-8')
    _df['Title'] = os.path.split(file)[-1]
    if _df.shape[0] > 6:               # rejecting songs which are under 6 lines
        df = df.append(_df)
```

```python
df_group = df.groupby('Title').agg({'lyrics': lambda x: " ".join(x)}).reset_index()
# Remove .txt extension from the Title column
df_group['Title'] = df_group['Title'].str.replace(r'.txt$', '')
pd.options.display.max_colwidth = 5000
df_group.head()
```

# Read the second dataset from the files

```python
#function to read one folder to 2d array. This is to read the songs in Hindi language
def readDataH(path,emotion):
    Data=[]
    lyrics=[]
    print(path)
    file_temp = [os.path.join(path,i) for i in sorted(os.listdir(path)) if os.path.isfile(os.path.join(path,i))]

    for f in file_temp:
        if not f.endswith('_E') and not "info" in f:
            try:
                filename = join(os.getcwd(),path, f)
                print(filename)
                with open(filename,'r') as file:
                    lyric=file.read()
                    lyrics.append(lyric)
            except:
                import traceback 
                traceback.print_exc()
                break
                
    with open(path+"info",'r') as file:
        i=0
        for line in file:
            row=line.split(':')
            row=[i.strip() for i in row]
            row.append(lyrics[i])
            Data.append(row)
            i+=1

    return Data
```

```python
AngH_Songs = readDataH("/Users/sanleypeter/Desktop/Main_Project/examples/Sentiment-Analysis-of-songs-by-lyrics/Hindi_Dataset/Angry/", 'A')
HapH_Songs = readDataH("/Users/sanleypeter/Desktop/Main_Project/examples/Sentiment-Analysis-of-songs-by-lyrics/Hindi_Dataset/Happy/", 'H')
SadH_Songs = readDataH("/Users/sanleypeter/Desktop/Main_Project/examples/Sentiment-Analysis-of-songs-by-lyrics/Hindi_Dataset/Sad/", 'S')
RelH_Songs = readDataH("/Users/sanleypeter/Desktop/Main_Project/examples/Sentiment-Analysis-of-songs-by-lyrics/Hindi_Dataset/Relaxed/", 'R')
SongsHindi = [AngH_Songs, HapH_Songs, SadH_Songs, RelH_Songs]
```

```python
SongsWordsTrain = [[], [], []]
polarity = 0
for i in range(4):
    for song in SongsHindi[i]:
        s = song[5]
        title = song[1]
        if i == 0 or i == 2:
          polarity = -1
        else:
          polarity = 1
        SongsWordsTrain[0].append(title)
        SongsWordsTrain[1].append(s)
        SongsWordsTrain[2].append(polarity)
```

```python
out = np.transpose(SongsWordsTrain)
lyrics2 = pd.DataFrame(out)
```

```python
#function to read one folder to 2d array
def readData(path,emotion):
    Data=[]
    lyrics=[]
    print(path)
    file_temp = [os.path.join(path,i) for i in sorted(os.listdir(path)) if os.path.isfile(os.path.join(path,i))]

    for f in file_temp:
        if f.endswith('_E') and not "info" in f:
            try:
                filename = join(os.getcwd(),path, f)
                with open(filename,'r') as file:
                    lyric=file.read()
                    lyrics.append(lyric)
            except:
                import traceback 
                traceback.print_exc()
                break

    with open(path+"info",'r') as file:
        i=0
        for line in file:
            row=line.split(':')
            row=[i.strip() for i in row]
            row.append(lyrics[i])
            Data.append(row)
            i+=1

    return Data
```

```python
AngH_Songs_E = readData("/Users/sanleypeter/Desktop/Main_Project/examples/Sentiment-Analysis-of-songs-by-lyrics/Hindi_Dataset/Angry/", 'A')
HapH_Songs_E = readData("/Users/sanleypeter/Desktop/Main_Project/examples/Sentiment-Analysis-of-songs-by-lyrics/Hindi_Dataset/Happy/", 'H')
SadH_Songs_E = readData("/Users/sanleypeter/Desktop/Main_Project/examples/Sentiment-Analysis-of-songs-by-lyrics/Hindi_Dataset/Sad/", 'S')
RelH_Songs_E = readData("/Users/sanleypeter/Desktop/Main_Project/examples/Sentiment-Analysis-of-songs-by-lyrics/Hindi_Dataset/Relaxed/", 'R')
SongsHindi_E = [AngH_Songs_E, HapH_Songs_E, SadH_Songs_E, RelH_Songs_E]
```

```python
lyrics_eng = []
for i in range(4):
    for song in SongsHindi_E[i]:
        s = song[5]
        lyrics_eng.append(s)
```
